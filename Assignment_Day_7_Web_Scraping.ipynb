Jadi pada assignment kali ini saya ingin melakukan scraping data menggunakan Requests dan Beautiful-Soup terkait populasi indonesia berdasarkan provinsi. adapun situs yang saya gunakan yaitu
"https://en.wikipedia.org/wiki/Demographics_of_Indonesia"

```python
#Import Libraries
import requests  # For making HTTP requests
from bs4 import BeautifulSoup  # For web scraping
import pandas as pd  # For data manipulation
```

```python
#Fetching Web Page

url = "https://en.wikipedia.org/wiki/Demographics_of_Indonesia"  # URL of the webpage to scrape
response = requests.get(url)  # Making a GET request to the URL
soup = BeautifulSoup(response.content, "html.parser")  # Parsing the HTML content of the webpage
print(soup.title.text)  # Printing the title of the webpage
```

```python
#Scraping and Processing Tables

tables = soup.find_all("table")  # Finding all tables on the webpage

dataframes = []  # Empty list to store dataframes
for table in tables:
    rows = table.find_all("tr")[1:]  # Skipping the first row (header)
    data = []  # Empty list to store row data
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]
        data.append(cols)
    df = pd.DataFrame(data)  # Creating a DataFrame from the scraped data
    dataframes.append(df)  # Appending the DataFrame to the list
```

```python
dataframes[2].to_csv("populasi_indonesia_berdasarkan_provinsi.csv", index=False, header=False)  # Saving the first DataFrame to a CSV file
```
